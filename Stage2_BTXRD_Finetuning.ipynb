## Command-line run (Windows / PowerShell)
Use backticks in PowerShell, or `^` in CMD. In PyCharm, prefer *Run Configuration* (see below).

```python
# CMD example:
!python mura.py ^
  --stage2_dir "C:\BTXRD - Copy\figshare_stage2_runs" ^
  --outdir "C:\BTXRD - Copy\paper_figs" ^
  --stage2_model "tf_efficientnetv2_s" ^
  --stage2_task "malignant_vs_normal" ^
  --img_size 224
```

## PyCharm Run Configuration
Run > Edit Configurations > + Python
- Script path: `...\mura.py`
- Parameters: `--stage2_dir ... --outdir ...` (same as above, **no** carets)
- Working directory: folder containing `mura.py`
